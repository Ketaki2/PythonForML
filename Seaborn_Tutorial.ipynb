## Seaborn Tutorial
Offers a lot of statistical tools
It is the backbone of Exploratory Data Analysis

When given a data set,
- Univariate Analysis -> Analysing one variable 
- Bivariate Analysis  -> Analyzing two variables and their relationship
Seaborn can help us extensively in doing the Univariate and Bivariate Analysis
- More than two features? Seaborn helps us to analyze if we have multiple features in our dataset through plots

### Distribution Plots (for numerical features):
- jointplot
- pairplot
- distplot

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore") # this will help you ignore the warnings
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Here tip -> dependent feature (Value dependent on independent features)
# total_bill,sex,smoker,day,time,size -> independent feature
```

## Correlation with Heatmap
- A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection to find the interrelationship between all the features(only for numerical features(int or float object) not for categorical features(object type))
- Correlation- value ranging from -1 to +1 (Coefficient Correlation)

```python
df.corr() # shows only numerical features
```

```python
df.dtypes
```

```python
# Positive correlation between x and y implies: When x increases, so does y e.g. Total bill and tip have a +ve correlation of 67%
# Negative correlation between x and y implies: When x increases, y decreases 
```

```python
sns.heatmap(df.corr())  # Helps to visualize the correlation matrix
```

```python
# Suppose two independent features have a correlation very close to 1,
# you can consider 1 independent feature in your analysis rather than taking both
```

### JoinPlot
Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

### Univariate Analysis

```python
x=tuple('tip')
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='hex') # kind=hex would give you hexagons in the graph
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='reg') 
# kind=reg will give a Probability Density Function
# based on the histogram and also give you the best-fit regression line
```

### PairPlot (pairwise jointplot)
A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.


```python
# Used when more than one independent feature( but only applicable for numerical)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

### DistPlot
A distplot plots a univariate distribution of observations. The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10) # Kernel Density Estimation
```

### Categorial Plots (for categorical features)
Analysis of the categorical data points

- countplot
- bar plot
- boxplot
- violinplot

### Count Plot
A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.It shows the counts of observations in each categorical bin using bars

```python
sns.countplot(x="time",data=df)
```

```python
df["time"].value_counts()
```

```python
sns.countplot('sex',data=df)
```

```python
df["sex"].value_counts()
```

```python
sns.countplot(y="time",data=df)
```

### Barplot 
- used for univariate analysis
- Similar to CountPlot but here you need to specify both x and y [One of these needs to be numerical variable] (Countplot->only x or y is needed)

```python
sns.barplot(x="smoker",y="total_bill",data=df)
```

### Boxplot
- Boxplots(box and whisker plot) are a measure of how well distributed the data in a data set is.
- This graph represents the minimum, maximum, median(second quartile), first quartile and third quartile in the data set.

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow',hue='smoker')
```

### Violin plot
- A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.
- Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
